```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the random seed for reproducibility
np.random.seed(0)

# Generate synthetic data
data_size = 100
x = np.random.normal(0, 1, data_size)
y = 2 * x + np.random.normal(0, 1, data_size)
z = np.random.normal(0, 1, data_size)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'z': z})
df.head(4)
```

```python
# Compute Pearson Correlation
pearson_corr = df.corr(method='pearson')
print("Pearson Correlation:\n", pearson_corr)

# Compute Spearman Correlation
spearman_corr = df.corr(method='spearman')
print("Spearman Correlation:\n", spearman_corr)

# Compute Kendall Tau Correlation
kendall_corr = df.corr(method='kendall')
print("Kendall Tau Correlation:\n", kendall_corr)

```

```python
# Visualize Correlations
#plt.figure(figsize=(10, 8))

# Plot Pearson Correlation
#plt.subplot(2, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation')
plt.savefig('PearsonCorr.png')
plt.show()

# Plot Spearman Correlation
#plt.subplot(2, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation')
plt.savefig('SpearmanCorr.png')
plt.show()

# Plot Kendall Tau Correlation
#plt.subplot(2, 2, 3)
sns.heatmap(kendall_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Kendall Tau Correlation')
plt.savefig('kendall_Tau.png')
plt.show()

plt.tight_layout()
plt.show()
```

# Point-Biserial Correlation 

The Point-Biserial correlation is a special case of the Pearson correlation used when one variable is continuous and the other is binary. Here's how you can compute the Point-Biserial correlation using Python:



Step 1: Generate Synthetic Data
We'll create a synthetic dataset with one continuous variable and one binary variable.

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(0)

# Generate synthetic continuous data
data_size = 100
continuous = np.random.normal(0, 1, data_size)

# Generate synthetic binary data
binary = np.random.choice([0, 1], size=data_size)

# Create a DataFrame
df = pd.DataFrame({'continuous': continuous, 'binary': binary})
```

```python
df.head(4)
```

Step 2: Compute Point-Biserial Correlation
You can use the pointbiserialr function from scipy.stats to compute the Point-Biserial correlation.

```python
from scipy.stats import pointbiserialr

# Compute Point-Biserial correlation
point_biserial_corr, p_value = pointbiserialr(df['binary'], df['continuous'])
print("Point-Biserial Correlation:", point_biserial_corr)
print("P-value:", p_value)
```

```python
# Compute Point-Biserial correlation
point_biserial_corr, p_value = pointbiserialr(df['binary'], df['continuous'])

# Plot the result
plt.figure(figsize=(10, 5))
sns.boxplot(x='binary', y='continuous', data=df)
plt.title(f'Point-Biserial Correlation: {point_biserial_corr:.2f} (p-value: {p_value:.4f})')
plt.xlabel('Binary Variable')
plt.ylabel('Continuous Variable')
plt.savefig('pointbiserial.png')
plt.show()
```

# Cramer's V 
Cramer's V is a measure of association between two nominal variables, providing a value between 0 and 1. 

```python

import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(42)

# Generate synthetic categorical data
data_size = 100
categories1 = np.random.choice(['A', 'B', 'C'], size=data_size)
categories2 = np.random.choice(['X', 'Y', 'Z'], size=data_size)

# Create a DataFrame
df_cramer = pd.DataFrame({'categories1': categories1, 'categories2': categories2})
df_cramer.head(4)
```

```python
import scipy.stats as stats

def cramers_v(confusion_matrix):
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1)*(r - 1)) / (n - 1))
    r_corr = r - ((r - 1) ** 2) / (n - 1)
    k_corr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((k_corr - 1), (r_corr - 1)))

# Create a contingency table
contingency_table = pd.crosstab(df_cramer['categories1'], df_cramer['categories2'])

# Compute Cramer's V
cramers_v_value = cramers_v(contingency_table.to_numpy())

# Compute Cramer's V
cramers_v_value = cramers_v(contingency_table.to_numpy())
print("Cramer's V:", cramers_v_value)
```

```python
# Plot the contingency table as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, cmap="Blues", cbar=False, fmt='d')
plt.title(f"Contingency Table Heatmap\nCramer's V: {cramers_v_value:.2f}")
plt.xlabel('Categories 2')
plt.ylabel('Categories 1')
plt.savefig('cramersV.png')
plt.show()
```

```python
import rpy2.robjects as ro
from rpy2.robjects.packages import importr
from rpy2.robjects import numpy2ri

# Enable the conversion between numpy and R objects
numpy2ri.activate()

# Load the R 'polycor' library
utils = importr('utils')
#utils.install_packages('polycor')
polycor = importr('polycor')

# Simulate ordinal data (e.g., Likert scale responses)
import numpy as np
data1 = np.random.choice([1, 2, 3, 4, 5], size=100)
data2 = np.random.choice([1, 2, 3, 4, 5], size=100)

# Convert data to R objects
r_data1 = ro.IntVector(data1)
r_data2 = ro.IntVector(data2)

# Calculate polychoric correlation in R
polychoric_corr = polycor.polychor(r_data1, r_data2)
print("Polychoric Correlation Coefficient:", polychoric_corr[0])



# Create a DataFrame
df = pd.DataFrame({'Variable 1': data1, 'Variable 2': data2})


# Create a contingency table
contingency_table = pd.crosstab(df['Variable 1'], df['Variable 2'])

# Assuming polychoric_corr[0] contains the polychoric correlation coefficient
polychoric_corr_coefficient = polychoric_corr[0]  # Replace with polychoric_corr[0] from R output in actual use

# Visualize the contingency table with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, cmap="Blues", cbar=True, fmt='d')
plt.title(f"Joint Frequency of Ordinal Variables\nPolychoric Correlation Coefficient: {polychoric_corr_coefficient:.2f}")
plt.xlabel('Variable 2')
plt.ylabel('Variable 1')
plt.savefig('polychoric_corr.png')
plt.show()
```

```python
import pingouin as pg
import pandas as pd
import numpy as np

# Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    'X': np.random.normal(0, 1, 100),
    'Y': np.random.normal(0, 1, 100),
    'Z': np.random.normal(0, 1, 100)
})

data.head(4)
```

```python
# Calculate partial correlation between X and Y, controlling for Z
partial_corr = pg.partial_corr(data=data, x='X', y='Y', covar='Z')

# Display the result
print("Partial Correlation between X and Y controlling for Z:")
print(partial_corr)
```

```python
# Extract the partial correlation coefficient value
partial_corr_value = partial_corr['r'].values[0]

# Create a scatter plot of X vs Y
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['X'], y=data['Y'], alpha=0.7)
plt.title(f"Scatter plot of X vs Y\nPartial Correlation (controlling for Z): {partial_corr_value:.2f}")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.savefig('partialcorr.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pointbiserialr

# Generate sample data
np.random.seed(42)
data_size = 100
continuous = np.random.normal(0, 1, data_size)
binary = np.random.choice([0, 1], size=data_size)

# Create a DataFrame
df = pd.DataFrame({'continuous': continuous, 'binary': binary})

# Calculate the Rank-Biserial (Point-Biserial) Correlation
rank_biserial_corr, p_value = pointbiserialr(df['binary'], df['continuous'])
print("Rank-Biserial Correlation (Point-Biserial Approximation):", rank_biserial_corr)

# Create a boxplot for visualization
plt.figure(figsize=(8, 6))
sns.boxplot(x='binary', y='continuous', data=df)
plt.title(f'Boxplot of Continuous Data by Binary Group\nRank-Biserial Correlation: {rank_biserial_corr:.2f}')
plt.xlabel('Binary Variable')
plt.ylabel('Continuous Variable')
plt.grid(True)
plt.savefig('rankbiserial.png')
plt.show()
```
