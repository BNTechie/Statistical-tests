```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the random seed for reproducibility
np.random.seed(0)

# Generate synthetic data
data_size = 100
x = np.random.normal(0, 1, data_size)
y = 2 * x + np.random.normal(0, 1, data_size)
z = np.random.normal(0, 1, data_size)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'z': z})

```

```python
# Compute Pearson Correlation
pearson_corr = df.corr(method='pearson')
print("Pearson Correlation:\n", pearson_corr)

# Compute Spearman Correlation
spearman_corr = df.corr(method='spearman')
print("Spearman Correlation:\n", spearman_corr)

# Compute Kendall Tau Correlation
kendall_corr = df.corr(method='kendall')
print("Kendall Tau Correlation:\n", kendall_corr)

```

```python
# Visualize Correlations
plt.figure(figsize=(10, 8))

# Plot Pearson Correlation
plt.subplot(2, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation')

# Plot Spearman Correlation
plt.subplot(2, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation')

# Plot Kendall Tau Correlation
plt.subplot(2, 2, 3)
sns.heatmap(kendall_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Kendall Tau Correlation')

plt.tight_layout()
plt.show()
```

# Point-Biserial Correlation 

The Point-Biserial correlation is a special case of the Pearson correlation used when one variable is continuous and the other is binary. Here's how you can compute the Point-Biserial correlation using Python:



Step 1: Generate Synthetic Data
We'll create a synthetic dataset with one continuous variable and one binary variable.

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(0)

# Generate synthetic continuous data
data_size = 100
continuous = np.random.normal(0, 1, data_size)

# Generate synthetic binary data
binary = np.random.choice([0, 1], size=data_size)

# Create a DataFrame
df = pd.DataFrame({'continuous': continuous, 'binary': binary})
```

Step 2: Compute Point-Biserial Correlation
You can use the pointbiserialr function from scipy.stats to compute the Point-Biserial correlation.

```python
from scipy.stats import pointbiserialr

# Compute Point-Biserial correlation
point_biserial_corr, p_value = pointbiserialr(df['binary'], df['continuous'])
print("Point-Biserial Correlation:", point_biserial_corr)
print("P-value:", p_value)
```

# Cramer's V 
Cramer's V is a measure of association between two nominal variables, providing a value between 0 and 1. 

```python

import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(42)

# Generate synthetic categorical data
data_size = 100
categories1 = np.random.choice(['A', 'B', 'C'], size=data_size)
categories2 = np.random.choice(['X', 'Y', 'Z'], size=data_size)

# Create a DataFrame
df_cramer = pd.DataFrame({'categories1': categories1, 'categories2': categories2})
```

```python
import scipy.stats as stats

def cramers_v(confusion_matrix):
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1)*(r - 1)) / (n - 1))
    r_corr = r - ((r - 1) ** 2) / (n - 1)
    k_corr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((k_corr - 1), (r_corr - 1)))

# Create a contingency table
contingency_table = pd.crosstab(df_cramer['categories1'], df_cramer['categories2'])

# Compute Cramer's V
cramers_v_value = cramers_v(contingency_table.to_numpy())
print("Cramer's V:", cramers_v_value)
```
